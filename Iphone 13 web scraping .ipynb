```python
import requests 
from bs4 import BeautifulSoup as bs
```

```python
url='https://www.amazon.in/Apple-iPhone-13-128GB-Midnight/product-reviews/B09G9HD6PD/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
page=requests.get(url)
page
```

```python
page.content
```

```python
soup=bs(page.content,'html.parser')
soup
```

```python
names= soup.find_all('span',class_="a-profile-name")[2:]
names
```

```python
len(names)
```

```python
import pandas as pd
df=pd.DataFrame(names,columns={'Customer Name'})
df
```

```python
df=df.drop_duplicates(ignore_index=True)
df
```

```python
r_title=soup.find_all('a',class_='review-title')
r_title
```

```python
len(r_title)
```

```python
review_title=[]
for i in range(0,len(r_title)):
    review_title.append(r_title[i].get_text()[1:-1])
review_title
```

```python
dates=soup.find_all('span',class_='review-date')[2:]
dates
```

```python
len(dates)
```

```python

review_date=[]
for i in range(0,len(dates)):
    review_date.append(dates[i].get_text().replace('Reviewed in India  on ',''))
review_date
```

```python
df['Review Title']= review_title
df['Review Date']= review_date
df
```

```python
stars= soup.find_all('span',class_='a-icon-alt')[5:]
stars
```

```python
len(stars)
```

```python
review_rating=[]
for i in range(0,len(stars)):
    review_rating.append(stars[i].get_text()[:])
review_rating
    
```

```python
df['Review Rating']=review_rating
df
```

```python
rev= soup.find_all('span',class_='a-size-base review-text review-text-content')[:]
rev
```

```python
len(rev)
```

```python
review_body= []
for i in range(0,len(rev)):
    review_body.append((rev[i]).get_text().replace('\n','')[:])
review_body
```

```python
len(review_body)
```

```python
df['Review']=review_body
df
```

```python
df.to_csv('IP13 Reviews.csv')
```
